```python
#sorting 1st dose uptake % in an ascending order
df= pd.DataFrame(vaccine_uptake_different_regions_df, columns=['BoroughName','1st_Dose_Uptake_%'])

df.sort_values(by=['1st_Dose_Uptake_%'])
```

```python
#sorting 2nd dose uptake % in an ascending order
df= pd.DataFrame(vaccine_uptake_different_regions_df, columns=['BoroughName','2nd_Dose_Uptake_%'])

df.sort_values(by=['2nd_Dose_Uptake_%'])
```

```python
#to check if there is any difference between 1st and 2nd dose
df_1st_dose= pd.DataFrame(vaccine_uptake_different_regions_df, columns=['BoroughName','1st_Dose_Uptake_%'])
df_2nd_dose= pd.DataFrame(vaccine_uptake_different_regions_df, columns=['BoroughName','2nd_Dose_Uptake_%'])
```

```python
df_1st_dose= pd.DataFrame(vaccine_uptake_different_regions_df, columns=['BoroughName','1st_Dose_Uptake_%'])
df_2nd_dose= pd.DataFrame(vaccine_uptake_different_regions_df, columns=['BoroughName','2nd_Dose_Uptake_%'])

#making vaccine dose columns to be in parallel
df_1st_dose['2nd_Dose_Uptake_%']=df_2nd_dose['2nd_Dose_Uptake_%']
#calculting the % difference between 1st dose and 2nd dose
df_1st_dose['% doses difference'] = np.where(df_1st_dose['1st_Dose_Uptake_%'] == df_2nd_dose['2nd_Dose_Uptake_%'],0, (df_1st_dose['1st_Dose_Uptake_%'] - df_2nd_dose['2nd_Dose_Uptake_%'])/df_1st_dose['1st_Dose_Uptake_%']*100)
rounded_df=df_1st_dose.round(decimals=2)

print(rounded_df)
```

```python
rounded_df.describe()
```

Summary so far, 
- Vaccine uptake in all boroughs have similar trend between 1st and 2nd dose.
- No significant difference (<10%) between 1st and 2nd dose vaccine uptake. 
- 2nd dose vaccine uptake is slightly less than 1st dose vaccine uptake consistent in all boroughs.


```python
#since analysis above shows that both dose uptake % are similar, average is taken to be categorised into different uptake level
vaccine1_uptake_different_regions_df=pd.DataFrame(vaccine_uptake_different_regions_df, columns=['BoroughName','1st_Dose_Uptake_%','2nd_Dose_Uptake_%'])

average_vaccine_uptake =vaccine1_uptake_different_regions_df.copy() #to allow it to be appended into the current dataset
average_vaccine_uptake['average'] = average_vaccine_uptake.mean(numeric_only=True, axis=1)
average_vaccine_uptake.to_csv("average_2 vaccines_uptake.csv")

```

```python
#To categorise data into different level, as all data are higher than 51% hence start from 50%
#51%-60%: very low; 61%-70%: low; 71%-80%: medium; 81%-90%:high ; 91%-100%: very high;

#categorising
bins = [51, 61, 71, 81, 91]
names = ['very low', 'low', 'medium', 'high', 'very high']
d = dict(enumerate(names))
average_vaccine_uptake['Uptake level']=np.vectorize(d.get)(np.digitize(average_vaccine_uptake['average'], bins))
average_vaccine_uptake
```

```python
#ethnic c data read
ethnic_df = pd.read_csv("C:/Users/Sin Nee/OneDrive/CFG Data science course/Project/ethnic_c.csv")
#changing name of Borough column to BoroughName to match with the one in vaccination
ethnic_df.rename(columns={'Borough':'BoroughName'}, inplace=True)
#changin name of Hackney to Hackney and City of London
ethnic_df['BoroughName']=ethnic_df['BoroughName'].str.replace('Hackney','Hackney and City of London')

ethnic_df
```

```python
#calculating mean of all_bame_prop, all_black_prop, pakistani_or_bangladeshi_prop, all_indian_prop
ethnic_bame_df = ethnic_df.groupby(['BoroughName'])['all_bame_prop'].mean()
ethnic_black_df = ethnic_df.groupby(['BoroughName'])['all_black_prop'].mean()
ethnic_packistani_bangladeshi_df = ethnic_df.groupby(['BoroughName'])['pakistani_or_bangladeshi_prop'].mean()
ethnic_indian_df = ethnic_df.groupby(['BoroughName'])['all_indian_prop'].mean()
ethnic_black_df
```
