Weekly data on cases, deaths, and vaccinations per London Borough.

```python
import pandas as pd
#file phe_cases_london_boroughs
vax_cases_london = pd.read_csv("https://raw.githubusercontent.com/VonnieC1/CFG-Final-Group-Project-/main/phe_cases_london_boroughs.csv")
vax_cases_london
#filter 2021
vax_cases_london['date'] = pd.to_datetime(vax_cases_london['date'], format= '%d/%m/%Y')
vax_cases_london_2021 = vax_cases_london.loc[(vax_cases_london['date'] >= '01/01/2021')& (vax_cases_london['date'] <= '31/12/2021')]
vax_cases_london_2021
#drop rows that are all empty
vax_cases_london_2021.dropna(how = 'all')
vax_cases_london_2021 
```

```python
import pandas as pd
#file phe_deaths_london_boroughs
deaths_london = pd.read_csv("https://raw.githubusercontent.com/VonnieC1/CFG-Final-Group-Project-/main/phe_deaths_london_boroughs.csv")
deaths_london
#filter 2021
deaths_london['date'] = pd.to_datetime(deaths_london['date'], format= '%Y/%m/%d')
deaths_london_2021 = deaths_london.loc[(deaths_london['date'] >= '2021-01-01')& (deaths_london['date'] <= '2021-12-31')]
deaths_london_2021
#drop rows that are all empty
deaths_london_2021.dropna(how = 'all')
deaths_london_2021
```

```python
import pandas as pd
#combine two files vaccines age london boroughs
vax_age_london1 = pd.read_excel("https://github.com/VonnieC1/CFG-Final-Group-Project-/blob/main/phe_vaccines_age_london_boroughs%202021a.xlsx?raw=true")
vax_age_london2 = pd.read_excel("https://github.com/VonnieC1/CFG-Final-Group-Project-/blob/main/phe_vaccines_age_london_boroughs%202021b.xlsx?raw=true")
vax_age_london = vax_age_london1.merge(vax_age_london2)
vax_age_london
#filter 2021
vax_age_london['date'] = pd.to_datetime(vax_age_london['date'], format= '%Y/%m/%d')
vax_age_london_2021 = vax_age_london.loc[(vax_age_london['date'] >= '2021-01-01')& (vax_age_london['date'] <= '2021-12-31')]
vax_age_london_2021
#drop rows that are all empty
vax_age_london_2021.dropna(how = 'all')
vax_age_london_2021
```
