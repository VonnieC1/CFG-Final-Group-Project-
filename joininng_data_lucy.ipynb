```python
import pandas as pd
import openpyxl

vaccinations = pd.read_excel('/Users/lucy/Library/Mobile Documents/com~apple~CloudDocs/Documents/CFG/Data/Homeworks/Project/Vaccination_uptake_percentage.xlsx')
vaccinations_firstdose = vaccinations.groupby(['BoroughName'])['1st_Dose_Uptake_%'].mean()
vaccinations_seconddose = vaccinations.groupby(['BoroughName'])['2nd_Dose_Updake_%'].mean()
vaccinations_firstdose_2 = vaccinations_firstdose.to_frame()
vaccinations_seconddose_2 = vaccinations_seconddose.to_frame()
vaccinations2 = vaccinations_seconddose_2.join(vaccinations_firstdose_2)
vaccinations2.to_csv('/Users/lucy/Library/Mobile Documents/com~apple~CloudDocs/Documents/CFG/Data/Homeworks/Project/Vaccination_uptake_averaged.csv')
```

```python
ethnic = pd.read_csv('/Users/lucy/Library/Mobile Documents/com~apple~CloudDocs/Documents/CFG/Data/Homeworks/Project/ethnic_c.csv', index_col = 'LocalAuthority')
ethnic
```

```python
ethnic_bame = ethnic.groupby(['LocalAuthority'])['all_bame_prop'].mean()
ethnic_black = ethnic.groupby(['LocalAuthority'])['all_black_prop'].mean()
ethnic_packistani_bangladeshi = ethnic.groupby(['LocalAuthority'])['pakistani_or_bangladeshi_prop'].mean()
ethnic_indian = ethnic.groupby(['LocalAuthority'])['all_indian_prop'].mean()
ethnic_bame
```

Joining data using the Local Authority column from ethnicity data and borough name from vaccinations  

```python
vaccinations_ethnic1 = vaccinations2.join(ethnic_bame)
vaccinations_ethnic2 = vaccinations_ethnic1.join(ethnic_black)
vaccinations_ethnic3 = vaccinations_ethnic2.join(ethnic_packistani_bangladeshi)
vaccinations_ethnic4 = vaccinations_ethnic3.join(ethnic_indian)
vaccinations_ethnic4.head()
```
