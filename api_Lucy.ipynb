```python
from requests import get
from json import dumps
import pandas_datareader as pdr

ENDPOINT = "https://api.coronavirus.data.gov.uk/v1/data"
AREA_TYPE = "region"
AREA_NAME = "London"

file_path = 'https://github.com/VonnieC1/CFG-Final-Group-Project-/tree/main/data'

filters = [
    f"areaType={ AREA_TYPE }",
    f"areaName={ AREA_NAME }"
    #f"date={ date }"
]

structure = {
    "date": "date",
    "name": "areaName",
    "code": "areaCode",
    "dailyCases": "newCasesByPublishDate",
    "dailyDeaths" : "newDeaths28DaysByDeathDate",
    "Vaccinations" : "newPeopleVaccinatedFirstDoseByVaccinationDate"
}


api_params = {
    "filters": str.join(";", filters),
    "structure": dumps(structure, separators=(",", ":")),
}


formats = [
    "csv"
]

for fmt in formats:
    api_params["format"] = fmt
    response = get(ENDPOINT, params=api_params, timeout=10)
    assert response.status_code == 200, f"Failed request for {fmt}: {response.text}"
    print(response.content.decode())
    with open('data.csv', 'w') as f:
        f.write(response.text)
```
